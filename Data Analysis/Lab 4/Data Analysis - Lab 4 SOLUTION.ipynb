# Question 2

```python
#QUESTION 2- Google Play Store
#import statistics Library

# ========================================================================
# a).
print()
print("Read the .csv file into python using the CSV library")
import statistics as st
import csv

#import library to draw the graphs
import matplotlib.pyplot as plt
import numpy as np


#imports text file as a string

my_file = open("googleplay.csv","r")
# Create a CSV Reader object for my_file
csv_reader = csv.reader(my_file)
```

```python
# ========================================================================
# b).
print()
print("Put the .csv file into a suitable 2D list")
# Use the list() function to return the data in the CSV reader as a 2D list
apps = list(csv_reader)
```

```python
# ========================================================================
# c).
print("Print out that list to the screen only print the first 20 rows and 8 attributes.")

#Print out first 20 instances and 8 attributes
for app in apps[:20]:
    for element in app[:8]:
        print(element, end="\t")
    print()
print()
print()


```

```python
# d). Cast the numeric columns to their appropriate numeric type.
for app in apps:
    app[2] = float(app[2])
    app[3] = int(app[3])
    app[5] = int(app[5])
    app[7] = float(app[7])
```

```python
# ========================================================================
# e).
print("Display the total free games installed versus the total paid games installed.")
installed_free = 0
installed_paid = 0
for app in apps:
    if app[6] == 'Paid':
        installed_paid += app[5]
    elif app[6] == 'Free':
        installed_free += app[5]
    else:
        print("App is neithee Free nor Paid: ", app)
print(f"Number of installs of free apps: {installed_free:14,}.")
print(f"Number of installs of paid apps: {installed_paid:14,}.")
```

```python
# ========================================================================
# f).
print("Show the above results in a bar chart.")

import numpy as np
import matplotlib.pyplot as plt

installs = ('Paid', 'Free')
# Get 2 evenly spaced values, could have hardcoded [0, 1], start at 0, stop before 2, in steps of 1
start = 0
stop = len(installs)
step = 1

x_positions = np.arange(start, stop, step)
print(f"x_positions: {x_positions}.") # Just so you can see them

usages = [installed_paid, installed_free]  # Effectivly the Y axis values

# Plot along the X-axis, the usage values, align the bar over the center (or edge) of the X-axis point, and select the color
plt.bar(x_positions, usages, align='center', color='red')
# Along the X-axis, at the calculated positions (0..5 in this case), put the catagorical values, i.e. the langages.
plt.xticks(x_positions, installs)
plt.ylabel('Installs')
plt.title('Free vs Paid Installs/Downloads')

plt.show()
```

```python

# ========================================================================
# g).
print("Show the above in a pie chart.")

# These are the variable you need in your pie chart
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Paid', 'Free'
slice_sizes = [installed_paid, installed_free]
explode = (0.1, 0)  # only "explode" the 1st slice (i.e. 'Paid')

# This code you copy and paste
figure, axes = plt.subplots()

# You may wish to adjust some of these paramers
axes.pie(slice_sizes,       # The wedge slices 
         explode=explode,   # See the 'explode' array variable above, 2nd one (Hogs) is exploded
         labels=labels,     # See the 'labels variable above.'
         autopct='%14.2f%%', # Floating point, precision and scale percentage, printed inside the sloce
         shadow=True,       # Causes a shadow 
         startangle=90)     # The angle by which the start of the pie is rotated, counterclockwise from the x-axis.

# This code you copy and paste
axes.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Paid vs Free App Installs/Downloads')
plt.show()

```

```python

# ========================================================================
# h).
print("Calculate and display the minimum, maximum and average number of installs for an app.")

from statistics import mean

installs = [app[5] for app in apps]
min_installs = min(installs)
max_installs = max(installs)
avg_installs = int(mean(installs))
print(f"Minimum Installs: {min_installs:14,}.")
print(f"Average Installs: {avg_installs:14,}.")
print(f"Maximum Installs: {max_installs:14,}.")
print()

```

```python
# ========================================================================
# i).
print("Graph the above as a bar chart")

names = ('Minimum', 'Average', 'Maximum')
# Get 3 evenly spaced values, could have hardcoded [0, 1, 2], start at 0, stop before 3, in steps of 1
start = 0
stop = len(names)
step = 1

x_positions = np.arange(start, stop, step)
print(f"x_positions: {x_positions}.") # Just so you can see them

usages = [min_installs, avg_installs, max_installs]  # Effectivly the Y axis values

# Plot along the X-axis, the usage values, align the bar over the center (or edge) of the X-axis point, and select the color
plt.bar(x_positions, usages, align='center', color='red')
# Along the X-axis, at the calculated positions (0..5 in this case), put the catagorical values, i.e. the langages.
plt.xticks(x_positions, names)
plt.ylabel('Installs')
plt.title('Downloads/Installs')

plt.show()

```

```python
# ========================================================================
# j).
print("Plot a histogram showing the number of installs using 6 bins.")


plt.xlabel('Age')
plt.ylabel('Number of apps')
plt.title('App Installs')
plt.hist(installs, bins=10)
```

### e)	Extract the prices and the points into 1D lists for all the wines in the dataset, and display the prices, unformated

```python
points = [wine[3] for wine in wines[1:]]
print("Points: ", points)
prices = [wine[4] for wine in wines[1:]]
print("Prices: ", prices)
```

### f)	Graph both of these on a line graph. 

```python
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Get evenly spaced values, start at 1, stop before 2, in steps of 1
start = 1
stop = len(prices) + 1
step = 1
x_positions = np.arange(start, stop, step)
print(x_positions)

plt.ylabel('Price / Points')
plt.title('Price vs Points')
plt.plot(x_positions, prices, label='Prices')  #<<<<<<<<<
plt.plot(x_positions, points, label='Points')  #<<<<<<<<<
#plt.show()
plt.legend()
```
