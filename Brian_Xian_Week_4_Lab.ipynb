**4.	How many total missing values in each colum**


```python
nan_count = permits.isna().sum()

print(nan_count)
```

**5. Print out the percentage of the missing values in the dataset**

```python
#permits.isnull().sum() / data.shape[0] * 100
percent_missing = permits.isnull().sum() * 100 / len(permits)
print(percent_missing)
```

**6.	How many missing values in the columns "Street Number Suffix" and "Zipcode" ?**

```python
street_number = permits['Street Number Suffix'].isna().sum()

print(street_number)
```

```python
zip_code = permits["Zipcode"].isna().sum()

print(zip_code)
```

**7.	Create new dataFrame named “pemits_dropped” that has all the column with empty values.**

**8.	Calculate all number of the dropped columns.**
