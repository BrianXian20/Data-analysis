# Exercises

Follow the instructions to recreate the plots using this data:

## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt**

```python
import matplotlib.pyplot as plt
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
# Creates blank canvas
fig = plt.figure()

# adding to an axis
ax = fig.add_axes([0,0,1,1])

# ploting the axis
ax.plot(x,y)

# setting the axis
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Creates blank canvas
fig = plt.figure()

# putting 2 axes on the canvas
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# setting up the first axes
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# setting up the second axes
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Show figure object
fig
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
# create figure
fig = plt.figure()

# adding two axes to the figure
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# plotting the axes
ax.plot(x,z)

# setting up the first Axis
ax.set_xlabel('X')
ax.set_ylabel('Z')

# plotting the second axis
ax2.plot(x,y)
# setting up the first Axis
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

# Show figure object
fig
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# the first axis (x,y) has the plots on the figure and the line color is blue and line is --
axes[0].plot(x,y,color="blue", lw=3, ls='--')

# the second axis (x,z) has the plots on the figure and the line color is red and the line is -
axes[1].plot(x,z,color="red", lw=3, ls='-')

# Show figure object
fig
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
# Empty canvas of 1 by 2 subplots and the figsize 12 and 2
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

# the first axis (x,y) has the plots on the figure and the line color is blue
# and line width 5 and the line is a straight line
axes[0].plot(x,y,color="blue", lw=5)

# setting up the axes for the first figure
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

# the second axis (x,z) has the plots on the figure and the line color is red
# and line width is 3 and the line is -
axes[1].plot(x,z,color="red", lw=3, ls='--')

# setting up the axes for the second figure
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
